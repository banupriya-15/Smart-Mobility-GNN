```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV file
df = pd.read_csv('co_accidents.csv')

# Step 2: Extract the year from Start_Time column
df['Year'] = pd.to_datetime(df['Start_Time']).dt.year

# Step 3: Count the number of accidents for each year
accidents_per_year = df['Year'].value_counts().sort_index()

# Step 4: Plot the distribution
plt.figure(figsize=(10, 6))
accidents_per_year.plot(kind='bar')
plt.title('Distribution of Accidents Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV file
df = pd.read_csv('co_accidents.csv')

# Step 2: Extract the year and month from Start_Time column
df['Year'] = pd.to_datetime(df['Start_Time']).dt.year
df['Month'] = pd.to_datetime(df['Start_Time']).dt.month

# Step 3: Group the data by year and month
accidents_per_year_month = df.groupby(['Year', 'Month']).size().unstack(fill_value=0)

# Step 4: Plot the distribution for each year by month
fig, ax = plt.subplots(figsize=(12, 8))
accidents_per_year_month.plot(kind='bar', stacked=True, ax=ax)
plt.title('Monthly Distribution of Accidents for Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.legend(title='Month', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
